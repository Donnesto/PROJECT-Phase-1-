# 2.Analysing tn_movie_budgets data

In this dataset analysis our focus is on understanding the financial aspects of movie production, specifically looking into production budgets, worldwide gross profits, and domestic profits. The main objectives are to identify movies with the highest gross profits both globally and domestically, and to gain insights into the distribution of movie profits.

i) Which movies has the highest Worldwide gross?

```python
tn_movie_budgets.sort_values(by="worldwide_gross", ascending=False)
tn_movie_budgets.head(10)
```

```python
# Barchart showing the Top Grossing Movies 
plt.figure(figsize=(18, 14))
top_movies = tn_movie_budgets.head(10)  
plt.bar(top_movies['movie'], top_movies['worldwide_gross'])
plt.xlabel('Movie')
plt.ylabel('Worldwide Gross')
plt.title('Top Performing Movies (Top 10)')
plt.xticks(rotation=45, ha='right')  
plt.show()
```

ii) Which movies made the highest gross profit world_wide and locally?

```python
# Convert 'production_budget' and 'worldwide_gross' to numeric values
tn_movie_budgets['domestic_gross'] = tn_movie_budgets['domestic_gross'].replace('[\$,]', '', regex=True).astype(float)
tn_movie_budgets['production_budget'] = tn_movie_budgets['production_budget'].replace('[\$,]', '', regex=True).astype(float)

# Calculate gross profit worldwide and locally
tn_movie_budgets['profit_worldwide'] = tn_movie_budgets['worldwide_gross'] - tn_movie_budgets['production_budget']
tn_movie_budgets['profit_domestic'] = tn_movie_budgets['domestic_gross'] - tn_movie_budgets['production_budget']

# Sort the dataset based on gross profit worldwide
sorted_worldwide = tn_movie_budgets.sort_values(by='profit_worldwide', ascending=False)

# Sort the dataset based on gross profit domestically
sorted_domestic = tn_movie_budgets.sort_values(by='profit_domestic', ascending=False)

# Retrieve the top movies
top_worldwide = sorted_worldwide.head(10) 
top_domestic = sorted_domestic.head(10)  

# Display the top movies
print("Top Movie Worldwide:")
print(top_worldwide[['movie', 'profit_worldwide']])

print("\nTop Movie Locally:")
print(top_domestic[['movie', 'profit_domestic']])

```

iii) Bar graph of top grossing movies and their profits worldwide

```python
#bar graph of top grossing movies and their profits worldwide
plt.figure(figsize=(18, 14))

plt.bar(top_worldwide['movie'], top_worldwide['profit_worldwide'], color='skyblue')
plt.xlabel('Movie')
plt.ylabel('Worldwide Profit (in billions)')
plt.title('Top Grossing Movies and Their Profits Worldwide')
plt.xticks(rotation=90)

plt.show()
```

```python
# Finding the relationship between a movies' production budget and how it sells world wide
tn_movie_budgets['production_budget'].corr(tn_movie_budgets['profit_worldwide'])
```

There is a positive relationship between a movies budget and the reception it gets worldwide
The higher the movie budget the high positive reception it gets.

```python
# Scatter plot of production budget vs. worldwide gross profit
plt.figure(figsize=(12, 8))
sns.scatterplot(x='production_budget', y='profit_worldwide', data=tn_movie_budgets, color='skyblue')
plt.title('Scatter Plot of Production Budget vs. Worldwide Profit')
plt.xlabel('Production Budget (in billions)')
plt.ylabel('Worldwide Profit (in billions)')
plt.show()
```

```python
#THIS IS FOR TN MOVIE BUDGETS TO SHOW GRAPH OF PROFIT 
# Remove non-numeric characters and convert to numeric
tn_movie_budgets['domestic_gross'] = tn_movie_budgets['domestic_gross'].replace('[\$,]', '', regex=True).astype(float)
tn_movie_budgets['production_budget'] = tn_movie_budgets['production_budget'].replace('[\$,]', '', regex=True).astype(float)

# Calculate profit
tn_movie_budgets['profit'] = tn_movie_budgets['domestic_gross'] - tn_movie_budgets['production_budget']

# Create a histogram of profit values 
plt.figure(figsize=(12, 8))
sns.histplot(tn_movie_budgets['profit'], bins=20, color='skyblue', kde=True)
plt.title('Distribution of Movie Profits')
plt.xlabel('Profit (Domestic Gross  - Production Budget) (in billions)')
plt.ylabel('Frequency')
plt.show()
```

Highlights the frequency of different profit ranges, making it easy to see where most movies fall in terms of profitability.
This provides a concise overview of profit distribution, making it easy to grasp the general landscape of movie profitability.

# 3.Analysing Bom.movie_gross data

This dataset will aid in analyzing the financial performance of movies, understanding the distribution of revenue across different markets, and identifying successful studios or movie genres based on their box office performance.

i) Top grossed movies 

```python

# Convert 'domestic_gross' and 'foreign_gross' to numeric values
bom_movie_gross['domestic_gross'] = bom_movie_gross['domestic_gross'].replace('[\$,]', '', regex=True).astype(float)
bom_movie_gross['foreign_gross'] = bom_movie_gross['foreign_gross'].replace('[\$,]', '', regex=True).astype(float)

# Calculate total gross revenue (domestic and foreign)
bom_movie_gross['total_gross'] = bom_movie_gross['domestic_gross'] + bom_movie_gross['foreign_gross']

# Sort the DataFrame by total gross revenue in descending order
top_movies = bom_movie_gross.sort_values(by='total_gross', ascending=False).head(10) 
```

```python

# Plotting the bar chart
plt.figure(figsize=(12, 8))
plt.bar(top_movies['title'], top_movies['total_gross'], color=['blue', 'blue', 'blue', 'red', 'blue', 'brown', 'blue', 'blue', 'blue', 'blue'])
plt.xlabel('Movie Title')
plt.ylabel('Total Gross Revenue (in billions)')
plt.title('Top Grossing Movies')
plt.xticks(rotation=45, ha='right')  
plt.show()
```

ii) Which studios generate highest gross

```python
# Group by studio and calculate total gross revenue for each studio
studio_high_gross = bom_movie_gross.groupby('studio')['total_gross'].sum()
studio_high_gross.sort_values(ascending=False) 
```

iii) Yearly trends of gross distribution 

```python


# Identify the highest-performing studio
top_studio = bom_movie_gross.groupby('studio')['total_gross'].sum().max()

# Filter data for the top studio
top_studio_data = bom_movie_gross[bom_movie_gross['studio'] == top_studio]


# Calculate average yearly gross for the top studio
avg_yearly_gross = top_studio_data.groupby('year')['total_gross'].mean().reset_index()
print(avg_yearly_gross)


```

```python
# Identify the top 10 performing studios
top_studios = bom_movie_gross.groupby('studio')['total_gross'].sum().max(10).index

# Filter data for the top 10 studios
top_studios_data = bom_movie_gross[bom_movie_gross['studio'].isin(top_studios)]

# Calculate average yearly gross for the top 10 studios
avg_yearly_gross = top_studios_data.groupby(['studio', 'year'])['total_gross'].mean().reset_index()

# Plotting the bar graph
plt.figure(figsize=(14, 10))
sns.barplot(x='year', y='total_gross', hue='studio', data=avg_yearly_gross, palette='viridis')
plt.xlabel('Year')
plt.ylabel('Average Yearly Gross Revenue (in billions)')
plt.title('Average Yearly Gross Revenue for Top 10 Studios Over the Years')
plt.legend(title='Studio', box_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

The graph helps in understanding the historical performance of the top studio, enabling Microsoft's stakeholders to make informed decisions about future investments, marketing strategies, or content creation based on past trends and patterns.

# 4.Analysing im.db data

 This dataset is used to understand top genres and also top directors as this will help Microsoft in knowing what kind of films have a higher reception among the audience also know what kind of people to collaborate with in this industry.

```python
#connecting to the database
conn = sqlite3.connect('im.db')
```

```python
# Selecting the table names
tables_name = """SELECT name 
                      AS 'Table Names' 
                      FROM sqlite_master 
                      WHERE type='table';"""

pd.read_sql(tables_name, conn)
```

i) How many genres do  we have?

```python
#how many genres do we have do we have
genres = """
   SELECT genres
     FROM movie_basics
 GROUP BY genres

;
"""
data = pd.read_sql(genres ,conn).dropna()
data.count()
```

ii) Top genres

```python
#Identify top genres
genre_ratings = """
   SELECT genres,avg(averagerating) AS average_ratings
     FROM movie_basics
     JOIN movie_ratings
    USING(movie_id)
 GROUP BY genres
 ORDER BY average_ratings DESC
 


;
"""
pd.read_sql(genre_ratings, conn).dropna()
```

iii) Most viewed genres

```python
#Identify most viewed genres
genre_ratings = """
   SELECT genres,sum(numvotes) AS People_watched
     FROM movie_basics
     JOIN movie_ratings
    USING(movie_id)
 GROUP BY genres
 ORDER BY people_watched desc

 


;
"""
pd.read_sql(genre_ratings, conn).dropna()
```

iv) Average movie length

```python
movie_length = """
  SELECT movie_id,avg(runtime_minutes) AS average_runtime,start_year
    FROM movie_basics
 GROUP BY start_year
 


 
    
;
"""
imdata = pd.read_sql(movie_length,conn)
imdata.dropna()
```

v) Best Directors

```python
#Total number of directors
directors_details = """
  SELECT person_id AS director_id, primary_name AS director_name
    FROM directors
    JOIN persons
   USING(person_id)
GROUP BY person_id
;
"""
imdata1 = pd.read_sql(directors_details,conn)
imdata1.info()
```

We have 109251 Directors

vi) Out of all this directors, which ones are highest rated? 

```python
highest_director_ratings = """
  SELECT person_id AS director_id, primary_name As director_name, COUNT(movie_id) AS movies_rated, avg(averagerating) AS director_movies_average_rating
    FROM directors
LEFT JOIN persons
USING(person_id)
    JOIN movie_ratings
    USING(movie_id)
GROUP BY director_id
HAVING movies_rated > 5
ORDER BY director_movies_average_rating DESC



 
    
;
"""
imdata2 = pd.read_sql(highest_director_ratings,conn)
imdata2
```
